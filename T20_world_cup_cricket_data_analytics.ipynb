# Data Preprocessing---üì•üñ®üñ®üñ®üñ®üñ®

```python
df['First Innings Score'] = pd.to_numeric(df['First Innings Score'],errors='coerce')
df['Fall of wickets First Innings'] = pd.to_numeric(df['Fall of wickets First Innings'],errors='coerce')
df['Second Innings Score'] = pd.to_numeric(df['Second Innings Score'],errors='coerce')
df['Fall of wickets Second Innings'] = pd.to_numeric(df['Fall of wickets First Innings'],errors='coerce')
df['Winning Margin'] = pd.to_numeric(df['Winning Margin'],errors='coerce')
df['Highest Score'] = pd.to_numeric(df['Highest Score'],errors='coerce')
df['Best Bowler Figure(Wickets Taken)'] =pd.to_numeric(df['Best Bowler Figure(Wickets Taken)'],errors='coerce')
df['Best Bowler Figure(Runs Recieved)'] = pd.to_numeric(df['Best Bowler Figure(Runs Recieved)'],errors='coerce')
```

# Complete Analysis On T20 World Cup ---- EDA üìäüìä

```python
df['Winners'].value_counts().plot(kind = 'bar',
                                  figsize = (16,6),
                                 color = 'Green',
                                 title = 'T20 WORLD CUP WINNING SCORE PER TEAM',
                                 ylabel='Number of Score',
                                 xlabel='Winner Teams')
```

```python
plt.figure(figsize = (16,6))
sns.countplot(data = df,
             x = df['Won by'],width=0.5)
plt.title('Winning Match By Wickets -- Runs -- Rain ')
plt.ylabel('Match Frequancy')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.countplot(data = df,
             x = df['Winning Margin'],)
plt.title('Winning Match By The Margine Score')
plt.ylabel('Match Frequancy')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sorted_scores = df['Highest Score'].sort_values(ascending=False)
top_10_scores = sorted_scores.head(10)
filtered_df = df[df['Highest Score'].isin(top_10_scores)]
sns.countplot(data=filtered_df, x='Highest Score')
plt.title('Top 10 Highest Score ')
plt.ylabel('Frequency of Score')
plt.show()
```

```python
plt.figure(figsize=(16,6))
df_1 = df[['Highest Score','Top Scorer']]
df_1 = df_1.sort_values(by='Highest Score',ascending= False)
sorted_values = df_1.head(10)
filter_10 = df_1[df_1[['Highest Score','Top Scorer']].isin(sorted_values)]
sns.barplot(data = filter_10,
           x = filter_10['Top Scorer'],
           y = filter_10['Highest Score'])
plt.title('Top 10 Batter In The World-Cup Whose Done Highest Score')
plt.xlabel('Batter Name')
plt.show()
```
